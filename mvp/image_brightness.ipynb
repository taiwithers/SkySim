```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
squares = 40
size = 10
buffer = 50
colours = np.linspace(1 / squares, 0.3, squares)

fig, ax = plt.subplots(figsize=(20, 1))

screen = np.zeros((size * squares + buffer * (squares + 1), buffer * 2 + size))
for i in range(squares):
    left = size * i + buffer * (i + 1)
    right = left + size
    top, bottom = buffer, buffer + size
    screen[left:right, top:bottom] = colours[i]
    ax.text(
        (left + right) / 2,
        top - buffer / 4,
        f"{colours[i]:.2f}",
        c=f"{colours[i]:.2f}",
        ha="center",
    )

ax.imshow(screen.T, vmin=0, vmax=1, cmap="Greys_r")
ax.set_frame_on(False)
ax.set(xticks=[], yticks=[])
plt.show()
```
